```python
import matplotlib.pyplot as plt
        
def open_file(filename: str) -> list:
    with open(f"results/{filename}.txt", "r") as file:
        data = [float(line) for line in file]
        return data
        
    
```

```python
indexs = [i for i in range(4, 100 + 1)]
result_annealing = open_file("result/sumulated_annealing_mod_population")
result_neibhor = open_file("result/nearest_neibhor_population")
")

fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))
ax.plot(indexs, result_annealing, color="green", linewidth=1.0, linestyle="-")
ax.plot(indexs, result_neibhor, color="orange", linewidth=1.0, linestyle="-")

ax.grid()
ax.set_xlabel(r' ', fontsize=12)
ax.set_ylabel(r'', fontsize=12)
ax.set_title("   ")
ax.legend(["simulated annealing", "nearest neibhor"])
plt.savefig("grafics/simulated and nearest both with population.png")
plt.show()
```
