```python
import matplotlib.pyplot as plt
        
def open_file(filename: str) -> list:
    with open(f"results/{filename}.txt", "r") as file:
        data = [int(line) for line in file]
        return data
        
    
```

```python
indexs = [i for i in range(4, 100 + 1)]
result_classic = open_file("classic")
result_kmethod = open_file("kmethod")
result_modKmethod = open_file('modKmethod')
result_popKmethod = open_file('popKmethod')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(8, 6))
ax.plot(indexs, result_classic, color="green", linewidth=1.0, linestyle="-")
ax.plot(indexs, result_kmethod, color="orange", linewidth=1.0, linestyle="-")
# ax.plot(indexs, result_modKmethod, color="blue", linewidth=1.0, linestyle="-")
# ax.plot(indexs, result_popKmethod, color="red", linewidth=1.0, linestyle="-")

ax.grid()
ax.set_xlabel(r'номер теста', fontsize=12)
ax.set_ylabel(r'сумма', fontsize=12)
ax.set_title("сравнение всех модификаций")
ax.legend(["Classic", "K method", "annefaling + k method", "population k method"])
plt.savefig("grafics/end.png")
plt.show()
```
